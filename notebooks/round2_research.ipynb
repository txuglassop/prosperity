```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices1 = pd.read_csv('./p3_data/round2/prices_round_2_day_-1.csv', sep=';')
prices2 = pd.read_csv('./p3_data/round2/prices_round_2_day_0.csv', sep=';')
prices3 = pd.read_csv('./p3_data/round2/prices_round_2_day_1.csv', sep=';')
prices = pd.concat([prices1, prices2, prices3])
```

```python
products_of_interest = ['CROISSANTS', 'DJEMBES', 'JAMS', 'PICNIC_BASKET1', 'PICNIC_BASKET2']
df = prices[prices['product'].isin(products_of_interest)]
```

```python
croissants = df[df['product'] == 'CROISSANTS']
djembes = df[df['product'] == 'DJEMBES']
jams = df[df['product'] == 'JAMS']
pb1 = df[df['product'] == 'PICNIC_BASKET1']
pb2 = df[df['product'] == 'PICNIC_BASKET2']
```

```python
diff1 = pb1['mid_price'].values - 6 * croissants['mid_price'].values - 3 * jams['mid_price'].values - djembes['mid_price'].values
diff2 = pb2['mid_price'].values - 4 * croissants['mid_price'].values - 2 * jams['mid_price'].values
```

```python
plt.figure(figsize=(10, 4))
plt.plot(diff1, label='diff1')
plt.plot(diff2, label='diff2')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)

plt.title('kys raph')
plt.legend()
plt.show()
```
