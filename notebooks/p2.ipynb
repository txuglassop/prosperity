```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(prices: list, product: str):
    for price in prices:
        asset = price[price['product'] == product]
        plt.figure(figsize=(15,5))
        plt.plot(asset['mid_price'])
        plt.show()
```

# Round 3

```python
prices1 = pd.read_csv('p2_data/round3/prices_round_3_day_0.csv', sep=';')
prices2 = pd.read_csv('p2_data/round3/prices_round_3_day_1.csv', sep=';')
prices3 = pd.read_csv('p2_data/round3/prices_round_3_day_2.csv', sep=';')
prices = [prices1, prices2, prices3]

for product in np.unique(prices1['product']):
    print(f'\n\nPRINTING PRICES FOR {product}')
    plot(prices, product)
```
